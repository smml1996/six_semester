## 1. Imágenes


```python
#Importando librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cargando una imágen en una escala de grises
img  = cv2.imread('cuzco.jpg',0) # el valor 0 indica escala de grises, sin ese valor, aparecería en colores
```

```python
# Para probar si la imagen fue cargada
print(img) # si cargó la imagen, deberá aparecer una matriz de valores
print(img.shape)
```

```python
# Para poder ver la imagen (no cerrar la imagen, solo darle ENTER)
cv2.imshow('image',img)
```

```python
# Para ver la imagen usando matplot
plt.imshow(img) #la opción cmap='gray' debería mostrar la imagen en escala de grises
plt.xticks([]), plt.yticks([]) #para ocultar tick valores en X y Y
plt.show()
```

```python
#cap = cv2.VideoCapture('http://192.168.1.102:4747/mjpegfeed')
cap = cv2.VideoCapture(0)
```

```python
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read() 
    if ret == False:
        print( "Error: No device found")
        break
    

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
